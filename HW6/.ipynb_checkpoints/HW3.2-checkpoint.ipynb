```python
import string
from sklearn.datasets import load_digits
import random
import csv
```

```python
with open("training-data.txt") as csvfile:
    lines=csv.reader(csvfile)
    dataset = list(lines)
trainingSet = []
C=[0] * 90
rate = 0.2
weight = [0.2]*5
for i in range(len(dataset)):
    if (i>=0 and i<45):
        trainingSet.append([1, float(dataset[i][0]), float(dataset[i][1]), float(dataset[i][2]), float(dataset[i][3])])
        C[i] = 0
    else:
        trainingSet.append([1, float(dataset[i][0]), float(dataset[i][1]), float(dataset[i][2]), float(dataset[i][3])])
        C[i] = 1
```

```python
with open("testing-data.txt") as csvfile:
    lines=csv.reader(csvfile)
    dataset = list(lines)
testSet = []
C_test=[0] * 10
for i in range(len(dataset)):
    if (i>=0 and i<5):
        testSet.append([1, float(dataset[i][0]), float(dataset[i][1]), float(dataset[i][2]), float(dataset[i][3])])
        C_test[i] = 0
    else:
        testSet.append([1, float(dataset[i][0]), float(dataset[i][1]), float(dataset[i][2]), float(dataset[i][3])])
        C_test[i] = 1
print(testSet)
print(C_test)
```

```python
def CH(pos, data):
    h = 0
    for i in range(5):
        h+=round(weight[i]*data[i], 1)
    h = round(h, 1)
    if h<=0:
        h = 0
    else:
        h = 1
    return (C[pos]-h)
```

```python
def Weight(ch, x):
    for i in range(5):
        weight[i] = round ((weight[i] + rate*ch*x[i]), 1)
```

```python
done = 0
count = 0
while done!=1 :
    print(count)
    for i in range(90):
        ch = CH(i, trainingSet[i])
        if (ch!=0):
            Weight(ch, trainingSet[i])
            count = 0
        else:
            count+=1
        if (count==90):
            done = 1
print(weight)
```
